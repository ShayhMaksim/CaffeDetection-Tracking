```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
words=["TrackerTLD","MedianFlow","TrackerBoosting","TrackerKCF","TrackerCSRT","TrackerMIL"]
```

```python
TrackerTLD=pd.read_csv("TrackerTLD")
TrackerMedianFlow=pd.read_csv("MedianFlow")
TrackerBoosting=pd.read_csv("TrackerBoosting")
TrackerKFC=pd.read_csv("TrackerKCF")
TrackerCSRT=pd.read_csv("TrackerCSRT")
TrackerMIL=pd.read_csv("TrackerMIL")
```

```python
TrackerBoosting.columns
```

```python
TrackerTLD0=TrackerTLD[TrackerTLD["try"]==0]
TrackerMedianFlow0=TrackerMedianFlow[TrackerMedianFlow["try"]==0]
TrackerBoosting0=TrackerBoosting[TrackerBoosting["try"]==0]
TrackerKFC0=TrackerKFC[TrackerKFC["try"]==0]
TrackerCSRT0=TrackerCSRT[TrackerCSRT["try"]==0]
TrackerMIL0=TrackerMIL[TrackerMIL["try"]==0]
```

```python
TrackerMedianFlow1=TrackerMedianFlow[TrackerMedianFlow["try"]==1]
TrackerCSRT1=TrackerCSRT[TrackerCSRT["try"]==1]
```

```python
plt.figure(figsize=(24,12))
plt.rc('axes', labelsize=20)
plt.rc('axes',titlesize=20)
plt.rc('legend', fontsize=20)
plt.title("График положения цели") # заголовок
plt.xlabel("X, пиксель") # ось абсцисс
plt.ylabel("Y, пиксель") # ось ординат
plt.grid()      # включение отображение сетки
plt.plot(TrackerMIL0['C_X'], TrackerMIL0['C_Y'],'p',label="TrackerMIL")  # построение графика
plt.plot(TrackerBoosting0['C_X'], TrackerBoosting0['C_Y'],'p',label="TrackerBoosting")
plt.plot(TrackerTLD0['C_X'], TrackerTLD0['C_Y'],'p',label="TrackerBoosting")
plt.legend(loc=0)
plt.savefig('График положения цели')
```

```python
plt.figure(figsize=(24,12))
plt.rc('axes', labelsize=20)
plt.rc('axes',titlesize=20)
plt.rc('legend', fontsize=20)
plt.title("График положения цели 1") # заголовок
plt.xlabel("X, пиксель") # ось абсцисс
plt.ylabel("Y, пиксель") # ось ординат
plt.grid()      # включение отображение сетки
plt.plot(TrackerKFC0['C_X'], TrackerKFC0['C_Y'],'p',label="TrackerKFC")  # построение графика
plt.plot(TrackerMedianFlow0['C_X'], TrackerMedianFlow0['C_Y'],'p',label="TrackerMedianFlow0")
plt.plot(TrackerCSRT0['C_X'], TrackerCSRT0['C_Y'],'p',label="TrackerCSRT")
plt.legend(loc=0)
plt.savefig('График положения цели 1')
```

```python
plt.figure(figsize=(24,12))
plt.rc('axes', labelsize=20)
plt.rc('axes',titlesize=20)
plt.rc('legend', fontsize=20)
plt.title("График положения цели 2") # заголовок
plt.xlabel("X, пиксель") # ось абсцисс
plt.ylabel("Y, пиксель") # ось ординат
plt.grid()      # включение отображение сетки
plt.plot(TrackerMedianFlow1['C_X'], TrackerMedianFlow1['C_Y'],'p',label="TrackerMedianFlow0")
plt.plot(TrackerCSRT1['C_X'], TrackerCSRT1['C_Y'],'p',label="TrackerCSRT")
plt.legend(loc=0)
plt.savefig('График положения цели 2')
```

```python
plt.figure(figsize=(24,12))
plt.rc('axes', labelsize=20)
plt.rc('axes',titlesize=20)
plt.rc('legend', fontsize=20)
plt.title("График fps") # заголовок
plt.xlabel("номер кадра") # ось абсцисс
plt.ylabel("fps") # ось ординат
plt.grid()      # включение отображение сетки
plt.plot(TrackerMIL['Unnamed: 0'], TrackerMIL['fps'],label="TrackerMIL")  # построение графика
plt.plot(TrackerBoosting['Unnamed: 0'], TrackerBoosting['fps'],label="TrackerBoosting")
plt.plot(TrackerTLD['Unnamed: 0'], TrackerTLD['fps'],label="TrackerBoosting")
plt.plot(TrackerKFC['Unnamed: 0'], TrackerKFC['fps'],label="TrackerKFC")  # построение графика
plt.plot(TrackerMedianFlow['Unnamed: 0'], TrackerMedianFlow['fps'],label="TrackerMedianFlow0")
plt.plot(TrackerCSRT['Unnamed: 0'], TrackerCSRT['fps'],label="TrackerCSRT")
plt.legend(loc=0)
plt.savefig('График fps')
```
